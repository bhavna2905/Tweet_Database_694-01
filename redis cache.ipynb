```python
#Top 10 users with the most followers worldwide (done)

MYSQL_HOST = "localhost"
MYSQL_USER = "root"
MYSQL_PASSWORD = "Admin"
MYSQL_DATABASE = "twitter_data_"

REDIS_HOST = "localhost"
REDIS_PORT = 6379
REDIS_DB = 0

def get_top_users_with_most_followers():
    r = redis.Redis(host=REDIS_HOST, port=REDIS_PORT, db=REDIS_DB)

    if r.exists("top_users_with_most_followers"):
        top_users = r.get("top_users_with_most_followers")
        top_users = json.loads(top_users)
        return top_users

    db = mysql.connector.connect(
        host=MYSQL_HOST, user=MYSQL_USER, password=MYSQL_PASSWORD, database=MYSQL_DATABASE
    )
    cursor = db.cursor()

    query = """
        SELECT screen_name, followers_count
        FROM users
        ORDER BY followers_count DESC
        LIMIT 10
    """
    cursor.execute(query)
    top_users = cursor.fetchall()

    top_users_dict = [{"screen_name": user[0], "followers_count": user[1]} for user in top_users]

    r.set("top_users_with_most_followers", json.dumps(top_users_dict))

    cursor.close()
    db.close()

    return top_users_dict

top_users = get_top_users_with_most_followers()
for user in top_users:
    print(f"{user['screen_name']}: {user['followers_count']} followers")
```

```python
#Top 10 verified users in the USA (done)
import mysql.connector
import redis
import json

db = mysql.connector.connect(user='root', password='Admin', host='localhost', database="twitter_data_")
cursor = db.cursor()

query = """
    SELECT screen_name, location FROM users
    WHERE verified = TRUE AND 
    (location = 'USA' or location = 'USA')
    ORDER BY followers_count DESC
    LIMIT 10
"""
cursor.execute(query) 

# Get the result
users = cursor.fetchall()
for row in users:
        print(row)

r = redis.Redis(host='localhost', port=6379, db=0)

for user in users:
    # Convert user data to JSON
    user_json = json.dumps(user)

    # Add user data to a Redis set
    r.sadd('verified_users_in_usa', user_json)

print("Data loaded into Redis.")


all_users = r.smembers('verified_users_in_usa')

users = [json.loads(user) for user in all_users]
print(len(users))
for user in users:
    print(user)
```

```python
#Users with the most friends worldwide (done)

db = mysql.connector.connect(user='root', password='Admin', host='localhost', database="twitter_data_")
cursor = db.cursor()

query = """
    SELECT screen_name, location, friends_count FROM users 
    ORDER BY friends_count DESC LIMIT 10
"""
cursor.execute(query) 

users = cursor.fetchall()
for row in users:
        print(row)

r = redis.Redis(host='localhost', port=6379, db=0)

for user in users:
    user_json = json.dumps(user)

    r.sadd('top_10_users_with_most_friends', user_json)

print("Data loaded into Redis.")

all_users = r.smembers('top_10_users_with_most_friends')

users = [json.loads(user) for user in all_users]
print(len(users))
for user in users:
    print(user)
```

```python
#Most followed non-verified accounts (done)

db = mysql.connector.connect(user='root', password='Admin', host='localhost', database="twitter_data_")
cursor = db.cursor()

query = """
    SELECT screen_name, location,followers_count  FROM users 
    WHERE verified = False
    ORDER BY statuses_count DESC LIMIT 10
"""
cursor.execute(query) 

users = cursor.fetchall()
for row in users:
        print(row)

r = redis.Redis(host='localhost', port=6379, db=0)

for user in users:
    user_json = json.dumps(user)

    r.sadd('Most_followed_non-verified_accounts', user_json)

print("Data loaded into Redis.")

all_users = r.smembers('Most_followed_non-verified_accounts')

users = [json.loads(user) for user in all_users]
print(len(users))
for user in users:
    print(user)
```

```python
#Top 10 users in a specific location (e.g., London, Tokyo) (done)

db = mysql.connector.connect(user='root', password='Admin', host='localhost', database="twitter_data_")
cursor = db.cursor()
location = 'London'  # Change this to the location you're interested in

query = f"""
   SELECT screen_name, location, followers_count FROM users 
   WHERE location LIKE '%{location}%'
   ORDER BY followers_count DESC LIMIT 10
"""
cursor.execute(query) 

users = cursor.fetchall()
for row in users:
        print(row)

r = redis.Redis(host='localhost', port=6379, db=0)

for user in users:
    user_json = json.dumps(user)

    r.sadd('Most_followed_non-verified_accounts', user_json)

print("Data loaded into Redis.")

all_users = r.smembers('Most_followed_non-verified_accounts')

users = [json.loads(user) for user in all_users]
print(len(users))
for user in users:
    print(user)
```
