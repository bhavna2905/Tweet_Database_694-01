```python
"""
Queries

Top 10 users with the most followers worldwide
Top 10 verified users in the USA  
Users with the most friends worldwide 
Top 10 users with the most tweets 
Users who have liked the most tweets 
Most followed non-verified accounts 
Top 10 users in a specific location (e.g., London, Tokyo) 
Most active users in terms of tweet frequency """
```

```python
MYSQL_HOST = "localhost"
MYSQL_USER = "root"
MYSQL_PASSWORD = "Admin"
MYSQL_DATABASE = "twitter_data_"

db = mysql.connector.connect(
    host=MYSQL_HOST,
    user=MYSQL_USER,
    password=MYSQL_PASSWORD,
    database=MYSQL_DATABASE
)

cursor = db.cursor()


#replace 'London' in the top_10_users_specific_location query with the actual location youâ€™re interested in
queries = {
    'top_10_most_followers': """
        SELECT id, name, screen_name, followers_count 
        FROM users 
        ORDER BY followers_count DESC 
        LIMIT 10
    """,
    'top_10_verified_users_USA': """
        SELECT id, name, screen_name, followers_count 
        FROM users 
        WHERE location LIKE '%USA%' AND verified = 1 
        ORDER BY followers_count DESC 
        LIMIT 10
    """,
    'most_friends': """
        SELECT id, name, screen_name, friends_count 
        FROM users 
        ORDER BY friends_count DESC 
        LIMIT 1
    """,
    'top_10_most_tweets': """
        SELECT id, name, screen_name, statuses_count 
        FROM users 
        ORDER BY statuses_count DESC 
        LIMIT 10
    """,
    'most_likes': """
        SELECT id, name, screen_name, favourites_count 
        FROM users 
        ORDER BY favourites_count DESC 
        LIMIT 1
    """,
    'most_followed_non_verified': """
        SELECT id, name, screen_name, followers_count 
        FROM users 
        WHERE verified = 0 
        ORDER BY followers_count DESC 
        LIMIT 1
    """,
    'top_10_users_specific_location': """
        SELECT id, name, screen_name, followers_count 
        FROM users 
        WHERE location LIKE '%London%' 
        ORDER BY followers_count DESC 
        LIMIT 10
    """,
    'most_active_users': """
        SELECT id, name, screen_name, statuses_count/(DATEDIFF(NOW(), created_date)+1) as tweet_frequency 
        FROM users 
        ORDER BY tweet_frequency DESC 
        LIMIT 1
    """
}

results = {}
for query_name, query in queries.items():
    cursor.execute(query)
    results[query_name] = cursor.fetchall()

with open('results.pkl', 'wb') as f:
    pickle.dump(results, f)

print("Results have been cached.")

db.close()


with open('results.pkl', 'rb') as f:
    cached_results = pickle.load(f)

for query_name, result in cached_results.items():
    print(f"\n{query_name}:")
    for row in result:
        print(row)
```
